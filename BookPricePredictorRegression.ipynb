```python
book=book.dropna()
```

```python
book.set_index('Author', inplace=True)
book=book.drop(columns=['Title','Edition'])
```

```python
two_hot=pd.get_dummies(book['Genre'])
# Drop column as it is now encoded
book = book.drop('Genre',axis = 1)
# Join the encoded df
book = book.join(two_hot)
```

```python
three_hot=pd.get_dummies(book['BookCategory'])
# Drop column as it is now encoded
book = book.drop('BookCategory',axis = 1)
# Join the encoded df
book = book.join(three_hot)
```
